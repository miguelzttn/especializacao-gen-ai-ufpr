9. ANÁLISE DE VENDAS
   Uma loja registra suas vendas diárias em uma lista. Cada venda é representada por uma 
   sublista contendo [nome_produto, quantidade, preco_unitario]. 
   Crie um programa que:
   - Calcule o faturamento total
   - Encontre o produto mais vendido (em quantidade)
   - Crie uma lista apenas com os produtos que tiveram faturamento acima de R$100,00
   - Mostre a média de preço dos produtos vendidos


```python

db = []

def registrar_venda(
    db: list, 
    nome_produto: str, 
    quantidade: int, 
    preco_unitario: float
) -> list[list[str | int | float]]:
    db.append([nome_produto, quantidade, preco_unitario])

def calcular_faturamento_total(db: list) -> float: 
    faturamento_por_vendas = [i[1] * i[2] for i in db]
    faturamento_total = sum(faturamento_por_vendas)
    return faturamento_total

def encontrar_produto_mais_vendido(db: list) -> str:
    return sorted(db, key=lambda x: x[1])[0]

def filtrar_por_faturamento(db: list, filtro: float) -> list:
    db_ordenada = sorted(db, key=lambda x: x[1] * x[2])
    db_ordenada = [i for i in db_ordenada if i[1] * i[2] >= filtro]
    return db_ordenada
    
def media_de_precos_vendidos(db: list) -> float:
    produtos = [i[0] for i in db]
    precos = [i[1] for i in db]
    quantidades = [i[2] for i in db]

    produtos_unicos = list(set(produtos))
    
    precos_unicos = [0 for _ in produtos_unicos]
    qtd_unicos = [0 for _ in produtos_unicos]


    for i, produto_unico in enumerate(produtos_unicos):
        for produto, preco, quantidade in zip(produtos, precos, quantidades):
            if produto_unico == produto:
                precos_unicos[i] += (preco * quantidade)
                qtd_unicos[i] += 1

    medias = []
    for preco, qtd in zip(precos_unicos, qtd_unicos):
        media = preco / qtd
        medias.append(media)

    return [produtos_unicos, medias]


registrar_venda(db, "Arroz", 1, 10.20)
registrar_venda(db, "Feijão", 2, 5.20)
registrar_venda(db, "Feijão", 2, 10)
registrar_venda(db, "Celular", 1, 1200)

print(100 * "-")
print("Banco de dados:\n")
print(db)
print(100 * "-" + "\n")

print("Calcular Faturamento Total", calcular_faturamento_total(db))
print("Pedidos Mais Vendidos", encontrar_produto_mais_vendido(db))
print("Faturamento >= 100", filtrar_por_faturamento(db, filtro=100.00))
print("Média de precos vendidos por produto", media_de_precos_vendidos(db))

```

11. LISTA ESPIRAL
    Crie um programa que gere uma matriz quadrada de tamanho N (fornecido pelo usuário) 
    preenchida com números em ordem crescente no formato espiral, começando do canto 
    superior esquerdo e girando no sentido horário.
    
    Exemplo para N=4:
    1   2   3   4
    12 13  14   5
    11 16  15   6
    10  9   8   7



```python

def gera_matriz_espiral(n: int) -> list[list[int]]:

    max_inter = 100

    X = [[0 for j in range(n)] for i in range(n)]
    x = 0

    i, j, d = 0, 0, 0

    for _ in range(max_inter):

        # >
        while j < (n - d):
            x += 1
            X[i][j] = x
            j += 1

        # v
        j -= 1
        i += 1
        while i < (n - d):
            x += 1
            X[i][j] = x
            i += 1

        # <
        i -= 1
        j -= 1
        while j > (0 + d):
            x += 1
            X[i][j] = x
            j -= 1

        # ^
        while i > (0 + d):
            x += 1
            X[i][j] = x
            i -= 1

        d += 1
        i += 1
        j += 1

        if x == (n * n):
            break
    
    return X

X = gera_matriz_espiral(n=5)
for z in X:
    print(z)
    
```
